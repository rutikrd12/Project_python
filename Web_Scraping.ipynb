### Web Scraping

```python
# importing required libraries
import requests
import bs4
from bs4 import BeautifulSoup as bs
import urllib
```

```python
link='https://www.flipkart.com/search?q=phone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
page = requests.get(link)
txt=page.content
```

```python
soup = bs4.BeautifulSoup(txt,"lxml")
soup
```

```python
title_tag=soup.select('title')
print(soup.title.string)
```

#### Extracting Price of the Product

```python
price=soup.find('div',class_="_30jeq3 _1_WHN1")
price.text
```

### Extracting the data of all the Mobiles available on the page.

```python
# creating list to store the data.
products=[]             
prices=[]                
ratings=[]             
ram_rom = []                             
display = []             
camera = []              
battery = []             
processor = []          
warranty = []           
```

```python
for data in soup.findAll('div',class_="_3pLy-c row"):
        names=data.find('div', attrs={'class':'_4rR01T'})
        price=data.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating=data.find('div', attrs={'class':'_3LWZlK'})
        specification = data.find('div', attrs={'class':'fMghEO'})
        for each in specification:
            col=each.find_all('li', attrs={'class':'rgWa7D'})
            ram_rom_ =col[0].text
            display_ = col[1].text
            camera_ = col[2].text
            Battery_ = col[3].text
            processor_ = col[4].text
            warranty_ = col[5].text   
       # Adding all elements in Lists
        products.append(names.text) 
        prices.append(price.text) 
        ratings.append(rating.text)
        ram_rom.append(ram_rom_)
        display.append(display_) 
        camera.append(camera_) 
        battery.append(Battery_)
        processor.append(processor_)  
        warranty.append(warranty_)  
```

```python
#printing the length of each list(To know how many values/product are present in the each list )
print(len(products))
print(len(prices))
print(len(ratings))
print(len(ram_rom))
print(len(camera))
print(len(battery))
print(len(processor))
print(len(warranty))
```

### Storing the data into the structured format in the Data Frame.

```python
import pandas as pd
df=pd.DataFrame({'Product Name':products,'prices':prices,'ratings':ratings,"ram_rom":ram_rom,
    'camera':camera,'battery':battery,'processor':processor,'warranty':warranty},index=range(1,25))
df
```
